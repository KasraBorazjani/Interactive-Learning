## Sample Script For HardRacetrack Environment

```python
import gym
from HardRacetrack.HardRacetrackEnv import HardRacetrackEnv

env = HardRacetrackEnv()
```

```python
import matplotlib.pyplot as plt
a = env.reset()

for i in range(20):
    action = env.action_space.sample()
    action = [0, 0.5]       # instead of some random action, you should develp an agent to make a wise decision each step.
    obs, reward, done, info = env.step(action)
    print(reward)
    env.render()

env.close()
```

```python
fig, axes = plt.subplots(ncols=4, figsize=(20, 5))
for i, o in enumerate(obs):
    plt.sca(axes[i])
    plt.imshow(o, cmap='gray')
    plt.title(i)
plt.show()
```
